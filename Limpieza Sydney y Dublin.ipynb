# ***EXTRACCIÓN DE DATOS***




**Dublin**

```python
#Tipo de propiedad Entire Rental Unit de host Simon
filtro1_dublin=dataDublin[(dataDublin["host_name"]=="Simon") & (dataDublin["property_type"]== "Entire rental unit")]
filtro1_dublin
```

```python
#Mejores Host calificados en Lucan, Ireland
filtro2_dublin=dataDublin[(dataDublin["host_response_rate"]=="100%") & (dataDublin["host_acceptance_rate"]== "100%") 
                          & (dataDublin["host_response_time"]== "within an hour") & (dataDublin["host_is_superhost"]== "t")
                          & (dataDublin["host_has_profile_pic"]== "t") & (dataDublin["host_identity_verified"]== "t")
                          & (dataDublin["host_location"]== "Lucan, Ireland")]
filtro2_dublin
```

```python
#Volver numérica la variable "price"
dataDublin["price"] = dataDublin["price"].str.replace('$', '', regex=False)
dataDublin["price"] = pd.to_numeric(dataDublin["price"], errors='coerce')

print(dataDublin["price"].head())
```

```python
#Casas o departamentos grandes en DUBLIN
filtro4_dublin=dataDublin[(dataDublin["room_type"]=="Entire home/apt") & (dataDublin["bathrooms"]>4) 
                          & (dataDublin["bedrooms"]>6)]
filtro4_dublin
```

```python
#Cuartos privados en condominios mejores calfiicados con precio menor a $100
filtro5_dublin=dataDublin[(dataDublin["review_scores_accuracy"]==5) & (dataDublin["review_scores_cleanliness"]==5) 
                          & (dataDublin["review_scores_checkin"]==5) & (dataDublin["review_scores_communication"]==5)
                          & (dataDublin["review_scores_location"]==5) & (dataDublin["review_scores_value"]==5) 
                          & (dataDublin["price"]<100) &(dataDublin["property_type"]=="Private room in condo")]
filtro5_dublin
```

# ***OUTLIERS***

***Dublin***

```python
df_dublin.info()
```

```python
#Seleccionamos solo las variables numéricas
df_dublin_numerico = df_dublin.select_dtypes(include=['number'])
df_dublin_numerico = df_dublin_numerico.drop(columns=['id', 'scrape_id', 'host_id'])
df_dublin_numerico.head()
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
df_dublin_numerico.plot(kind='box', vert=False)
plt.title("Valores Atípicos cuantitativas dublin")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=df_dublin_numerico
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
outliersdata1= df_dublin_numerico[(y<=Limite_Superior)&(y>=Limite_Inferior)]
outliersdata1.head()
```

```python
#Corroboramos valores nulos del dataframe
valores_nulosdata1=outliersdata1.isnull().sum()
valores_nulosdata1
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_cleandata1=outliersdata1.copy()
data_cleandata1=data_cleandata1.fillna(round(outliersdata1.mean(),1))
data_cleandata1.head()
```

```python
#Corroboramos valores nulos del dataframe
valores_nulosdata1=data_cleandata1.isnull().sum()
valores_nulosdata1
```

```python
# Crear un nuevo DataFrame con las columnas específicas
df_dublin_categoricas = df_dublin[['host_response_time', 'host_acceptance_rate', 'property_type', 'room_type', 'amenities', 'host_is_superhost']]
df_dublin_categoricas.head()
```

```python
Dublin = pd.concat([df_dublin_categoricas,data_cleandata1], axis=1)
Dublin.head()
```
